```python
import torch
inputs = torch.tensor(
[[0.43, 0.15, 0.89], # Your (x^1)
[0.55, 0.87, 0.66], # journey (x^2)
[0.57, 0.85, 0.64], # starts (x^3)
[0.22, 0.58, 0.33], # with (x^4)
[0.77, 0.25, 0.10], # one (x^5)
[0.05, 0.80, 0.55]] # step (x^6)
)
```

# Calculate attention score matrix:

```python
attention_score_matrix = inputs @ inputs.T
attention_score_matrix
```

# Calculate attention weight matrix:

```python
attention_weight_matrix = torch.softmax(attention_score_matrix, dim = -1)
attention_weight_matrix
```

# Calculate context vector matrix:

```python
context_vector_matrix = attention_weight_matrix @ inputs
context_vector_matrix
```

**Note:** the context vectors, or even the input vectors (embedded one) are lying on the row, they are in the row space of the matrix.
